## Are there product categories with high sales but low average transaction amounts?

```python
hs=df.groupby("services",as_index=False)["products_used"].value_counts().sort_values(by="count",ascending=False).head(1)
hss=hs["products_used"].values[0]
hss
```

```python
lt=df.groupby("products_used",as_index=False)["t_amt"].sum().rename(columns={"t_amt":"total_t_amt"}).min()
ltt=lt["products_used"]
if(hss!=ltt):
    print("there is no product categories with high sales but low average transaction amounts")
else:
    print("there is a product category with high sales but low average transaction amounts",ltt)   
```
